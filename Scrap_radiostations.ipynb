```python
from bs4 import BeautifulSoup

import requests
import urllib.request
import time
url = 'https://wiki.ubuntuusers.de/Internetradio/Stationen'
```

```python
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

# print (soup.findAll('a'))
print (soup.prettify())
```

```python
cnt = 0
for ele in soup.find_all('pre', class_='notranslate'):
    print (cnt, ele.prettify())
    cnt += 1
```

```python
cnt = 0
for ele in soup.find_all('pre', class_='notranslate'):
    print ('\n', cnt, '\n', ele.string)
    cnt += 1
```

```python
cnt = 0
for ele in soup.find_all('pre', class_='notranslate'):
    print ('\n ======= ', cnt, '==========\n', ele.parent)
    print ('\n', cnt, '\n', ele.string)
    cnt += 1
```
